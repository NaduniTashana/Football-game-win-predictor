```python
rf.fit(train[predictors],train["target"])
```

```python
preds= rf.predict(test[predictors])
```

```python
grouped_matches= matches.groupby("team")
```

```python
#get previous 3 weeks performance and predict next week performances

def rolling_averages(group,cols,new_cols):
    group=group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean() #closed='left means not get the predict raw(last raw)
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
#apply this for all teams
matches_rolling= matches.groupby("team").apply(lambda x: rolling_averages(group, cols, new_cols))
```

```python
matches_rolling
```

```python
#drop extra index level that has team name
matches_rolling= matches_rolling.droplevel("team")
```

```python
matches_rolling
```

```python
#indexing correction to 0 to 210
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
#Retraing the model
def make_predictions(data, predictors):
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision= make_predictions(matches_rolling, predictors + new_cols)
```
